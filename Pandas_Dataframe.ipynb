```python
# Q4. Indexing using .loc
import pandas as pd

# Sample DataFrame with 25 rows
data = {
    "Product": ["Laptop", "Mobile", "Tablet", "Headphones", "Camera"] * 5,
    "Region": ["North", "South", "East", "West", "Central"] * 5,
    "Sales": [25000, 18000, 12000, 8000, 15000] * 5,
    "Quantity": [50, 70, 40, 90, 30] * 5
}
df = pd.DataFrame(data)

# 1. Display row with index 10
print(df.loc[10])

# 2. Display Sales value of row with index 20
print(df.loc[20, "Sales"])
```

```python
# Q5. Indexing using .iloc
import pandas as pd

# Sample DataFrame
data = {
    "Product": ["Laptop", "Mobile", "Tablet", "Headphones", "Camera", "Printer"],
    "Region": ["North", "South", "East", "West", "Central", "North"],
    "Sales": [25000, 18000, 12000, 8000, 15000, 10000],
    "Quantity": [50, 70, 40, 90, 30, 20]
}
df = pd.DataFrame(data)

# 1. First 6 rows and first 4 columns
print(df.iloc[:6, :4])

# 2. Sales value from the 3rd row
print(df.iloc[2, 2])
```

```python
# Q6. Row Slicing
import pandas as pd

# Sample DataFrame with 30 rows
data = {
    "Product": ["Laptop", "Mobile", "Tablet", "Headphones", "Camera"] * 6,
    "Region": ["North", "South", "East", "West", "Central"] * 6,
    "Sales": [25000, 18000, 12000, 8000, 15000] * 6,
    "Quantity": [50, 70, 40, 90, 30] * 6
}
df = pd.DataFrame(data)

# 1. Display first 10 rows
print(df.head(10))

# 2. Display rows 15 to 25
print(df.iloc[15:26])
```

```python
# Q7. Filtering Data
import pandas as pd

# Sample DataFrame
data = {
    "Product": ["Laptop", "Mobile", "Tablet", "Headphones", "Camera", "Printer"],
    "Region": ["North", "South", "East", "West", "East", "North"],
    "Sales": [25000, 18000, 12000, 8000, 2000, 1000],
    "Quantity": [50, 70, 40, 90, 30, 20]
}
df = pd.DataFrame(data)

# 1. Filter rows where Region = "East"
print(df[df["Region"] == "East"])

# 2. Filter rows where Sales > 1500
print(df[df["Sales"] > 1500])
```

```python
# Q8. Multiple Condition Filtering
import pandas as pd

# Sample DataFrame
data = {
    "Product": ["Laptop", "Mobile", "Tablet", "Headphones", "Camera", "Printer"],
    "Region": ["North", "South", "West", "West", "East", "West"],
    "Sales": [25000, 18000, 12000, 8000, 2000, 1500],
    "Quantity": [50, 70, 40, 90, 30, 20]
}
df = pd.DataFrame(data)

# Filter rows where Region is West AND Sales > 1200
filtered = df[(df["Region"] == "West") & (df["Sales"] > 1200)]
print(filtered)
```

```python
# Q9. Descriptive Statistics
import pandas as pd

# Sample DataFrame
data = {
    "Product": ["Laptop", "Mobile", "Tablet", "Headphones", "Camera", "Printer"],
    "Region": ["North", "South", "East", "West", "East", "North"],
    "Sales": [25000, 18000, 12000, 8000, 2000, 1000],
    "Quantity": [50, 70, 40, 90, 30, 20]
}
df = pd.DataFrame(data)

# Generate summary statistics
stats = df.describe()
print(stats)

# Extract required values
mean_sales = stats.loc["mean", "Sales"]
max_quantity = stats.loc["max", "Quantity"]
min_sales = stats.loc["min", "Sales"]

print("Mean Sales Value:", mean_sales)
print("Maximum Quantity Sold:", max_quantity)
print("Minimum Sales Value:", min_sales)
```

```python
# Q10. Aggregation Analysis
import pandas as pd

# Sample DataFrame
data = {
    "Product": ["Laptop", "Mobile", "Tablet", "Headphones", "Camera", "Printer"],
    "Region": ["North", "South", "East", "West", "East", "North"],
    "Sales": [25000, 18000, 12000, 8000, 2000, 1000],
    "Quantity": [50, 70, 40, 90, 30, 20]
}
df = pd.DataFrame(data)

# 1. Total sales
total_sales = df["Sales"].sum()

# 2. Average sales
average_sales = df["Sales"].mean()

# 3. Maximum sales
max_sales = df["Sales"].max()

# 4. Minimum sales
min_sales = df["Sales"].min()

print("Total Sales:", total_sales)
print("Average Sales:", average_sales)
print("Maximum Sales:", max_sales)
print("Minimum Sales:", min_sales)
```
